```python
import cv2
import matplotlib.pyplot as plt
import os
from PIL import Image 
import PIL 

```

```python
folder_dir = "./Edge-Samples" 
for images in os.listdir(folder_dir):
 
    # check if the image ends with png
    if (images.endswith(".jpg")):
        print(images)
        # creating a image object (main image) 
        os.chdir("/content/gdrive/MyDrive/Canny-Edge-Detection/Edge-Samples")
        img = cv2.imread(images)
        edges = cv2.Canny(img, 100, 200, 3, L2gradient=True)
        os.chdir("/content/gdrive/MyDrive/Canny-Edge-Detection/Edge-Samples/output")
        plt.imsave(images, edges, cmap='gray', format='png')
```
